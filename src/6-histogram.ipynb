### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Step 2: Line Chart in the first subplot
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values")
axs[0, 0].grid(True)

# Step 3: Bar Chart in the second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Step 4: Scatter Plot in the third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='green')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")

# Step 5: Pie Chart in the fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title("Programming Language Usage")

# Step 6: Histogram as an additional subplot
fig_hist, ax_hist = plt.subplots(figsize=(6, 4))
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
ax_hist.hist(data, bins=5, color='purple', edgecolor='black')
ax_hist.set_title("Histogram")
ax_hist.set_xlabel("Value")
ax_hist.set_ylabel("Frequency")

# Step 7: Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
